```python
import cv2
import matplotlib.pyplot as plt

# Load the image
image = cv2.imread('/content/task7_partA.jpg')

# Apply Non-Local Means Denoising
denoised_image = cv2.fastNlMeansDenoisingColored(image, None, 15, 15, 10, 21)

# Convert the image from BGR to RGB (Matplotlib expects RGB images)
denoised_image_rgb = cv2.cvtColor(denoised_image, cv2.COLOR_BGR2RGB)
cv2.imwrite('denoised_image.jpg', denoised_image)

# Display the denoised image using Matplotlib
plt.imshow(denoised_image_rgb)
plt.axis('off')  # Turn off axis numbers
plt.show()
```
